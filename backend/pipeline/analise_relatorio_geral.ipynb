```python
import pandas as pd

path = "/root/Documents/python_projects/estoque_farmacia_sesau/dados/Toda a rede.xls"

df = pd.read_excel(path, header=None)

# print(df)

estabelecimentos =df[df[1].astype(str).str.contains('Estabelecimento', case=True, na=False, regex=True)].dropna(axis=1, how='all').reset_index(drop=True)

estabelecimentos = estabelecimentos[1].str.split(': ', n=1, expand=True)
estabelecimentos = estabelecimentos.drop(columns=0)

print(estabelecimentos)
```
