```python
colunas = df.iloc[3].dropna().to_dict()
```

```python
df = df.rename(columns=colunas)
df
```

```python
estabelecimentos =df[df["Produto"].astype(str).str.contains('Estabelecimento', case=True, na=False, regex=True)].dropna(axis=1, how='all')#.reset_index(drop=True)

# estabelecimentos = estabelecimentos[1].str.split(': ', n=1, expand=True)
# estabelecimentos = estabelecimentos.drop(columns=0)
estabelecimentos['indice']=estabelecimentos.index
estabelecimentos = estabelecimentos.reset_index(drop=True)

estabelecimentos
```

```python
from numpy import nan
estab = estabelecimentos.groupby('Produto')['indice'].agg(['min'])
estab = estab.reset_index()
estab['max'] = estab['min'].shift(-1).fillna(len(df)).astype(int)
```

```python
estab
```

```python
estab = estab.rename(columns={'min': 'inicio', 'max': 'fim'})
estab['estabelecimento'] = estab['Produto'].str.split(': ', n=1, expand=True)[1]
estab = estab.drop(columns='Produto')

estab
```

```python
import os
silver_path = base_path + '/silver'
os.makedirs(silver_path, exist_ok=True)

# Define a linha de início dos dados reais no DataFrame original
LINHA_INICIO_DADOS_ORIGINAL = 4

for index, row in estab.iterrows():
    idx_inicio = row['inicio']
    idx_fim = row['fim']
    nome_estabelecimento = row['estabelecimento']

    # Fatia o DataFrame original para o estabelecimento atual
    df_estab = df.iloc[idx_inicio:idx_fim].copy()

    # Ajusta o início do fatiamento para ignorar linhas de cabeçalho/metadados no df_estab
    idx_dados_relativo = max(0, LINHA_INICIO_DADOS_ORIGINAL - idx_inicio)
    df_estab = df_estab.iloc[idx_dados_relativo:].copy()

    # Renomeia as colunas usando o dicionário 'colunas'
    df_estab = df_estab.rename(columns=colunas)

    # Remove colunas e linhas que são inteiramente NaN
    df_estab = df_estab.dropna(axis=1, how='all')
    df_estab = df_estab.dropna(axis=0, how='all')

    # Saneia o nome do estabelecimento para criar um nome de arquivo válido
    nome_sanitizado = nome_estabelecimento.replace('/', '_').replace(':', '_').replace('\\', '_').replace('*', '_').replace('?', '_').replace('"', '_').replace('<', '_').replace('>', '_').replace('|', '_')

    # Constrói o caminho e salva o DataFrame em um arquivo Excel na pasta 'silver'
    caminho_final = os.path.join(silver_path, f"{nome_sanitizado}.xlsx")
    df_estab.to_excel(caminho_final, index=False)

print("Arquivos Excel criados com sucesso no diretório 'silver'.")
```

# Extraindo informações de cada unidade

```python
#carregar arquivos da pasta silver

dataframe = pd.read_excel()
dataframe
```

Extraindo dados dos medicamentos

```python
medicamentos_df = dataframe[dataframe['Produto'].astype(str).str.contains('BR', case=True, na=False, regex=True)].dropna(axis=1, how='all').reset_index(drop=True)
medicamentos_df = medicamentos_df.rename(columns={'Produto': 'nome_medicamento'})
medicamentos_df[['codigo_medicamento','nome_medicamento']] = medicamentos_df['nome_medicamento'].str.split(' ', n=1, expand=True)


display(medicamentos_df)
```

```python
quantidade_df = dataframe[dataframe['Produto'].astype(str).str.contains('Total', case=True, na=False, regex=True)].dropna(axis=1, how='all').reset_index(drop=True)

quantidade_df = quantidade_df.drop(columns=['Produto',18])
quantidade_df = quantidade_df.rename(columns={12: 'quantidade'})

display(quantidade_df)
```

```python
estoque = pd.merge(medicamentos_df, quantidade_df, left_index=True, right_index=True, how='inner')

estoque = estoque[['codigo_medicamento','nome_medicamento','quantidade']]
estoque
```

```python
# save estoue in a json named estoque_processado_alves_pereira
estoque.to_json(base_path + '/estoque_processado_alves_pereira.json', orient='records')
# # SAVA A CSV WITH THE SAME NAME
# estoque.to_csv(base_path + '/estoque_processado_alves_pereira.csv', index=False)
```
